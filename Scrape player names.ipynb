```python
import urllib
import xml.etree.ElementTree as ET
from bs4 import BeautifulSoup, SoupStrainer
from datetime import timedelta, date
import os
import pickle
import pandas as pd
```

```python
def get_url(year, month, day):
        
    return 'http://gd2.mlb.com/components/game/mlb/year_' + year + '/month_' + month + '/day_' + day + '/'

def daterange(start_date, end_date):
    for n in range(int ((end_date - start_date).days)):
        yield start_date + timedelta(n)
```

```python
#Scrape all data from given date range according to wind conditions
year = 2015

if year == 2017:
    start_date = date(2017, 4, 2)
    end_date = date(2017, 10, 3)
elif year == 2018:
    start_date = date(2018, 3, 29)
    end_date = date(2018, 10, 2)
elif year == 2016:
    start_date = date(2016, 4, 3)
    end_date = date(2016, 10, 3)
elif year == 2015:
    start_date = date(2015, 4, 5)
    end_date = date(2015, 10, 6)
else:
    start_date = date(year, 3, 28)
    end_date = date(year, 10, 6)
    
if not os.path.isdir(str(year) + '_xmls'):
    os.mkdir(str(year) + '_xmls')
    
#players = pd.DataFrame(columns=['first_name', 'last_name', 'id'])

count = 0
for single_date in daterange(start_date, end_date):
    
    d = single_date.strftime("%Y-%m-%d").split('-')
    day_url = get_url(d[0], d[1], d[2])
    day_url = day_url[:-1]

    print(day_url)
    with urllib.request.urlopen(day_url) as f:
        webpage = f.read().decode('utf-8')
    soup = BeautifulSoup(webpage, 'html.parser')  
    a_list = soup.find_all('a')
    game_dirs = [ele['href'] for ele in a_list if 'gid' in ele.text]

    for gdir in game_dirs:
        gdir = gdir[6:]
        #print(gdir)
        try:
            with urllib.request.urlopen(day_url + gdir + 'players.xml') as f:
                p_text = f.read().decode('utf-8')
                
            xml_name = 'players_xmls/' + gdir[1:-1] + '_players.xml'
            savefile = open(xml_name, 'w')
            savefile.write(game_string)
            savefile.close()
            
            p_parsed = BeautifulSoup(p_text, 'xml')
            
            for p in p_parsed.findAll('player'):
                if int(p['id']) in players['id'].values:
                    continue
                players = players.append({'first_name': p['first'], 
                                'last_name': p['last'], 
                                'id': int(p['id'])}, ignore_index=True)
                
        except:
            try:
                with urllib.request.urlopen(day_url + gdir + 'linescore.xml') as f:
                    game_string = f.read().decode('utf-8')
                    soup = BeautifulSoup(game_string, 'xml')
                    if not soup.findAll('game')[0]['status'] == 'Postponed':
                        print('Something is wrong with ' + day_url + gdir)
            except:
                print("something very wrong with " + day_url + gdir)
                    
    print(len(players))
```
